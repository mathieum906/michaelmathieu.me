```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

unesco_url = 'http://whc.unesco.org/en/list/xml'
xml_data = requests.get(unesco_url).content

unesco_df = pd.DataFrame()

i = 0

for row in ET.XML(xml_data).getchildren():
    for col in row.getchildren():
        unesco_df.loc[i, col.tag] = col.text
    i += 1
```
